# **13 principle of clean code**
1. Don't Repeat Yourself (DRY): means Avoiding code duplication.
2. Single Responsibility Principle (SRP): Each component should be responsible for a single task or functionality.
3. Open/Closed Principle (OCP): Software entities should be open for extension but closed for modification.
4. Liskov Substitution Principle (LSP): derived classes should be able to substitute their base classes seamlessly.
5. Interface Segregation Principle (ISP): it is better to create smaller and more specific interfaces for the need of each client.
6. Dependency Inversion Principle (DIP):  DIP promotes depending on abstractions rather than concrete implementations.
7. Law of Demeter (LoD): an object should have limited knowledge about other objects.
8. Composition over Inheritance: Composition provides flexibility and avoids Inheritance limitations.
9. Keep Functions Small: Functions should be focused on a single task.
10. Meaningful Naming: the name should refer to the function of the named member.
11. Consistent Formatting: including identation,spaces and it increases code readability.
12. Unit Testing: To verify the the correctness of Individual components.
13. Continuous Refactoring: improving the design and structure of code without changing its task.

# **Sample techniques**
1. Random Sampling: Random sampling is a widely used technique that ensures each member of a population has an equal chance of being selected for the sample.
2. Stratified Sampling: Stratified sampling involves dividing the population into subgroups based on certain characteristics.
3. Cluster Sampling: Cluster sampling involves dividing the population into clusters or groups and then randomly selecting a few clusters to include in the sample.
4. Convenience Sampling: Convenience sampling involves selecting individuals who are readily available and accessible for inclusion in the sample.
5. Systematic Sampling: Systematic sampling involves selecting every nth individual from a population after randomly determining a starting point

# **Outliers**

```python
# how to detect and remove outliers
a = [1,2,3,4,5,7,8,2,6,1,9,6,7,2,110]
a.sort()
```

```python
q1 = int(0.25*(len(a)+1))
q3 = int(0.75*(len(a)+1))
iqr = int(a[q3]-a[q1])
min = int(a[q1]-(1.5*iqr))
max = int(a[q3]+(1.5*iqr))
for i in a:
  if (i < min) | (i > max):
    print(i)
```

```python
data3 = {
    "values":[1,2,1,2,1,50]
}
data3 = pd.DataFrame(data3)
data3 = pd.concat([data,data3],axis=1)
data3
```

```python
Q1 = data3["values"].quantile(0.25)
Q3 = data3["values"].quantile(0.75)
iqr = Q3 - Q1
lbound = Q1-(1.5*iqr)
ubound = Q3 + (1.5*iqr)
```

```python
outliers = data3[~((data3["values"] > lbound) & (data3["values"] < ubound))]
no_outliers = data3[(data3["values"] > lbound) & (data3["values"] < ubound)]
outliers
no_outliers
```
