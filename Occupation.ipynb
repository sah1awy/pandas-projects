### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user).

```python
import io
import requests
link = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
s = requests.get(link).content
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv(io.StringIO(s.decode('utf-8')),sep='|',on_bad_lines='skip')
users.sample(3)
```

### Step 4. Discover what is the mean age per occupation

```python
g = users.groupby(by="occupation")
g.aggregate(np.mean)["age"]
```

```python
g["age"].aggregate(np.mean)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
a = pd.crosstab(index=users.occupation,columns=users.gender)["M"]
b = pd.crosstab(index=users.occupation,columns=users.gender)["F"] + pd.crosstab(index=users.occupation,columns=users.gender)["M"]
(a/b*100).sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
g["age"].max()
```

```python
g["age"].min()
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
pd.crosstab(index=users.occupation,columns=users.gender,values=users.age,aggfunc=np.mean)
```

### Step 8.  For each occupation present the percentage of women and men

```python
g = users.groupby(["occupation","gender"]).count()
g = g["user_id"]
e = users.groupby('occupation').count()["user_id"]
print(e)
```
